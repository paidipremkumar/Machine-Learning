# Import datascience Libraries

```python
import numpy as np
import pandas as pd
import nltk #Natural language Tool Kit
```

## Step-1 :Cleaning Text
** Convertin into lower case letters and removing special characters and numerics**

```python
text = 'I like this machine learning class,because I like the concept of @Image Processing,@Python,#NLP and $Deep Learning. My class is till 5:30PM.'
```

```python
text
```

> Step-01: Converting into Lower Case

```python
text = text.lower()
text
```

> Step-02: Removing Special Characters

```python
import re #Regular expression
```

```python
text_clean = re.sub('[^a-zA-Z]',' ',text)
print(text)
print('After Clean')
print(text_clean)
```

## Step-2 :Stemming
**Convert string into list**


**Apply for loop and identify root word**

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
# Converting into list
text_list = text_clean.split()
```

```python
root_word = []
for words in text_list:
    word = ps.stem(words)
    root_word.append(word)
```

**Step-2: Removing Stopwords**

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words()
```

```python
stop = stopwords.words('english')
```

```python
stop.remove('and')
```

```python
stop.append('tea')
```

```python
stop
```

```python
# Removing Stopwords

clean = []
for words in root_word:
    if not words in stop:
        clean.append(words)
```
